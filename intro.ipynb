```python
# prompt: 컬럼 2개, 관측치 10개 짜리 데이터를 표준정규분포에서 추출해서 데이터 프레임에 넣어줘

import pandas as pd
import numpy as np

# 표준정규분포에서 10개의 관측치를 갖는 2개의 컬럼 생성
data = np.random.randn(10, 2)

# 데이터프레임 생성
df = pd.DataFrame(data, columns=['col1', 'col2'])

# 데이터프레임 출력
df
```

- 두 변수를 산포도로 시각화 (shift + enter)

```python
import matplotlib.pyplot as plt

# 산포도 그리기
plt.figure(figsize=(6, 4))
plt.scatter(df['col1'], df['col2'], alpha=0.7)
plt.xlabel('col1')
plt.ylabel('col2')
plt.title('Scatter Plot of col1 vs col2')
plt.grid(True)

# 그래프 표시
plt.show()
```

- col1 -> x1
- col2 -> x2
- title: "x1과 x2의 산포도"
